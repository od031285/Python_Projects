```python
import pandas as pd
from pathlib import Path

schools = pd.read_csv(Path("schools.csv"))
schools.head()
```

```python
#Finding schools with the best math scores
best_math_schools=schools[schools['average_math']>=600]
best_math_schools.head()
```

```python
# Identifying the top 10 performing schools
schools['total_SAT']=schools['average_math']+	schools['average_reading']+schools['average_writing']
top_10=schools.nlargest(10,'total_SAT').reset_index(drop=True)
top_10
```
